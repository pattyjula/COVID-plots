```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams
import numpy as np
import matplotlib.ticker as mtick
```

```python
df = pd.read_csv("Global_Mobility_Report.csv", parse_dates=['date'])#, dayfirst=True)
df.set_index('date')

# run a query for Los Angeles county
df.query('country_region_code == "US" and sub_region_1 == "California" and sub_region_2 == "Los Angeles County"',inplace = True)
df.head(2)
```

```python
sns.set(style="whitegrid")

plt.xticks(rotation=45)

#sns.set_style('darkgrid')
#fig, ax = plt.subplots()
ax = sns.lineplot(x="date", y="retail_and_recreation_percent_change_from_baseline", data=df, palette ="Set2")#, color= 'Set2')
rcParams['figure.figsize'] = 11.7,8.27
plt.title("LA County - retail and recreation percent change from baseline", size = 12)
source_txt = """Source: Google
https://www.google.com/covid19/mobility/
Last updated: """
last_date = df["date"].max()
source_txt += last_date.strftime("%Y-%m-%d")
plt.figtext(0.02, 0.57, source_txt,transform=ax.transAxes,
        bbox=dict(facecolor='white', alpha=0.8))
plt.show()
```
